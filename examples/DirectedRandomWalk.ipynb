```python
import math
```

creating a spiral with the below equation.

```python
k = [np.exp(0.01*x)*(math.cos(x) + 1j*math.sin(x)) for x in np.linspace(0,20*np.pi,500)]
```

If we plot RandomWalk with zero step size, we will be able to print out the path of the underlying curve we are supplying

```python
MotionPlot().plot(RandomWalk.RandomWalk(500,num_directions=5,path=k,step_size=0).get_motion_path())
```

Now trying some different step sizes in random walk along with the directed path that we supplied

```python
MotionPlot().plot(RandomWalk.RandomWalk(500,num_directions=5,path=k,step_size=0.01).get_motion_path())
```

```python
MotionPlot().plot(RandomWalk.RandomWalk(500,num_directions=5,path=k,step_size=0.05).get_motion_path())
```
